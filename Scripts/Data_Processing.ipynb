```python
import pandas as pd
```

```python
db = pd.read_csv(filepath_or_buffer="filtered_output.tsv", sep="\t")

db
```

```python
coding_del_order = ["frameshift_variant", "stop_gained", "splice_donor_variant", "splice_acceptor_variant",
                       "inframe_deletion", "inframe_insertion", "missense_variant", "start_lost", "stop_lost", "synonymous_variant"]


cons = list(db["Consequence"])
cons_simple = []
for index, con in enumerate(cons):
    final_string = con
    if "," in con:
        split_cons = con.split(",")
        final = 99
        for val in split_cons:
            if val in coding_del_order:
                rank = coding_del_order.index(val)
                if rank < final:
                    final = rank
        if final != 99:
            final_string = coding_del_order[final]
        else:
            final_string = None
    

    if final_string in coding_del_order:
        cons_simple.append(final_string)
    else:
        cons_simple.append(None)

print(cons_simple)

db['Consequence_simple'] = cons_simple
```
