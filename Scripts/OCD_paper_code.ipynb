```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
supplementals = pd.read_excel('/Users/emarshman/Research/Rotation_WangCapra/20260121_OCDCTD_tables.xlsx', sheet_name=None)
```

```python
pc1 = supplementals['S3 - Singletons']['PC1']
pc2 = supplementals['S3 - Singletons']['PC2']
ancestry_pred = supplementals['S3 - Singletons']["ancestry_pred"]


set(ancestry_pred)
```

```python

# 1. Define the mapping
color_map = {'AFR': 0, 'AMR': 1, 'EAS': 2, 'EUR': 3, 'SAS': 4, 'uncertain': 5}

# 2. Convert categories to a list of colors
colors = ancestry_pred.map(color_map)

cmap = plt.cm.get_cmap('Set1', 6) # 2 categories
cmap.set_bad('black') # Color for NaNs

fig, ax = plt.subplots()
val = ax.scatter(pc1, pc2, c=colors, cmap=cmap)
fig.colorbar(val) 
plt.show()
```
